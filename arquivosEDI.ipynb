```python
def parse_line(filename, pos, line, layout):
    parsed_data = {}
    start = 0
    parsed_data["nome_arquivo"] = filename
    parsed_data["posicao_linha"] = pos
    for field_name, field_length in layout:
        parsed_data[field_name] = line[start:start + field_length].strip()
        start += field_length
    return parsed_data

def read_fixed_width_files(directory, layouts):
    all_data = []
    
    for filename in os.listdir(directory):
        if filename.endswith(".txt"):
            pos_line = 0
            with open(os.path.join(directory, filename), 'r') as file:
                for line in file:
                    pos_line += 1
                    line_type = line[:2]
                    if line_type in layouts:
                        layout = layouts[line_type]
                        parsed_line = parse_line(filename, pos_line, line, layout)
                        all_data.append(parsed_line)
                    else:
                        print(f"Tipo de linha desconhecido: {line_type}")
    return all_data
```

```python
# Diretório contendo os arquivos de texto
directory = '/mnt/edi_proc'

# Ler e processar os arquivos
data = read_fixed_width_files(directory, layouts)

#print(f'Diretorio: {directory}')
#print(f'Layouts: {layouts}')

# Exibir os dados lidos
for record in data:
    print(record)
```
